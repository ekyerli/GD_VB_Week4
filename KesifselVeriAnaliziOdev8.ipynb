```python
#(1) Create a variable that contains the weighted average of the grades in the dataset. The number of students in the fourth grade is different from that of the eighth grade. So you will need a weighted average!4
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
from scipy.stats.mstats import winsorize
warnings.filterwarnings('ignore')
import seaborn as sns

states = pd.read_csv("states_all.csv")
```

```python
states=states.fillna(states.mean())
```

```python
#(2) What is the correlation between the variable you just created and the types of expenditures? Which expenditure item has more correlation than others?
weight0 = 0

for i in states.columns[-4:]:
    print(f"{i} değişkenindeki not-null değer sayısı:", states[i].notnull().sum())
    weight0 += states[i].notnull().sum()
print("\nToplam değer sayısı:", weight0)
```

```python
weight1 = states["AVG_MATH_4_SCORE"].notnull().sum() / weight0
weight2 = states["AVG_MATH_8_SCORE"].notnull().sum() / weight0
weight3 = states["AVG_READING_4_SCORE"].notnull().sum() / weight0
weight4 = states["AVG_READING_8_SCORE"].notnull().sum() / weight0

states["WEIGHTED_SCORE"] = ((weight1 * states["AVG_MATH_4_SCORE"]) + 
                               (weight2 * states["AVG_MATH_8_SCORE"]) + 
                               (weight3 * states["AVG_READING_4_SCORE"]) + 
                               (weight4 * states["AVG_READING_8_SCORE"]))
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

states_df = pd.concat([states2.iloc[:,5:10], states2.iloc[:, -1]], axis=1)

states_df.head()
states_df.corr()
```

```python
X=StandardScaler().fit_transform(states_df)
pca=PCA(n_components=len(states_df.columns))
principalComponents=pca.fit_transform(X)

exp_var=pca.explained_variance_ratio_*100
cumsum_var=np.cumsum(np.round(pca.explained_variance_ratio_,decimals=4)*100)
print(exp_var)
print("--"*31)
print(cumsum_var)
```
