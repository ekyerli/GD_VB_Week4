# Keşifsel Veri Analizi

# 1. Keşifsel veri analizi nedir?
Keşifsel veri analizi, veri bilimi boru hattının ilk yapı taşıdır. Modelleme aşamasına geçmeden önce, verilerimizdeki ilişkileri anlamamız gerekir. Verilerimizi ne kadar iyi anlarsak, modellerimizin en iyi performanslarını elde etmelerine yardımcı olacak en kullanışlı özellikleri daha başarılı bir şekilde belirleriz.

EDA'yı üç aşamadan veya adımdan oluşan yinelemeli bir süreç olarak kavramsallaştırırız. Aşağıdaki şekil bu aşamaları ve yinelemenin nasıl gerçekleştiğini göstermektedir:

1)Veri temizleme: EDA sürecinin ilk aşamasıdır. Veri kümesini alırız ve verilerde tespit edebileceğimiz sorunları gidermeye çalışırız.

2)Veriyi keşfetme: Bu ikinci adım, farklı özellikler ile verilerdeki modeller arasındaki ilişkileri keşfetmek için verilerin araştırılmasıdır. Bu aşamada kullandığımız temel araçlar, istatistikler ve görselleştirme teknikleridir. Araştırma sırasında verilerdeki yeni sorunlar tespit edersek, temizleme aşamasına geri döner ve yeni bulunan sorunları ortadan kaldırarak verileri temizleriz.

3)Özellik mühendisliği: EDA'nın son adımı, en kullanışlı özellikleri seçtiğimiz veya mevcut özelliklerden yeni özellikler oluşturduğumuz özellik mühendisliğidir. Bu aşama bizi veri bilimi boru hattının sonraki adımlarına götürecektir. Ancak verilerle ilgili bazı sorunlar tespit edersek, veri temizleme aşamasına başvurmalı ve verileri tekrar temizlemeliyiz. Ayrıca, mevcut özellikleri dönüştürürsek veya yenilerini yaratırsak, bu yeni özelliklerin kullanışlı olup olmadığını test etmek için veri keşfi aşamasına geri dönmemiz gerekir.

## 1. Sürekli değişkenler

Sürekli değişkenler, sonsuz sayıda potansiyel farklı değer alabilen değişkenlerdir temsil eder. Bu değişken türü muhtemelen gerçek hayattan veri kümeleriyle çalışırken karşılaşacağınız en yaygın veri türüdür. Hız, mesafe, ağırlık, boy, harcama vb. gibi birçok değer, sürekli değişkenlerle gösterilebilecek ölçümlerden bazılarıdır.

İki tür sürekli değişken vardır:

### 1.1. Aralık değişkenleri
Aralık değişkenleri, hem sırayı hem de farkı gösterir. Sıcaklık, aralık değişkeni için iyi bir örnektir. 10°C ila 15°C arasındaki fark 20°C ila 25°C derece (5°) arasındaki farkla aynıdır. Aralık değişkenleri ile ilgili en önemli nokta, mutlak bir sıfır noktası bulunmamasıdır. Örneğin, 0°C sıcaklığı, havada hiç sıcaklık olmadığı anlamına gelmez! Mutlak bir sıfır noktasının olmaması, örneğin, 20°C değişkenli bir aralıktan mantıksal olarak oran yapamayacağımız anlamına gelir. Örneğin 20°C, 10°C'nin iki katı sıcak değildir.

Aralık değişkenleriyle ilgili akılda tutulması gereken bir başka şey de, kategorik değişkenler olarak ele alınabilecek olmalarıdır. Bununla birlikte, doğal sürekli değişken olarak mutlak bir sıfır noktasına sahip olan bazı modellerin yaptığı varsayımla rahatsak, aralık değişkenlerini sürekli olarak değerlendirebiliriz.

### 1.2. Oran değişkenleri
İkinci sürekli değişken türü oran değişkenleri'dir. Sırayı, mesafeyi ve anlamlı bir mutlak sıfır değerini gösterir - bir oran değişkeni 0 puanına sahip olduğunda, değişken tarafından ölçülen miktarın hiçbirinin mevcut olmadığını biliriz. Mutlak bir sıfır değerinin varlığı, oran ve aralık değişkenleri arasındaki ana ayrımdır. Örnek olarak, yaş bir oran değişkenidir: 0 yaşında biri henüz doğmamıştır ve 20 yaşında biri 10 yaşında birinin iki katı yaşamıştır.

## 2. Kategorik değişkenler

Kategorik değişkenler kategorileri veya grupları temsil eder. Tanım olarak, kategorik bir değişken yalnızca sınırlı sayıda farklı değer alabilir. Sürekli ve kategorik değişkenler arasındaki temel ayrım noktası, bir değişkenin alabileceği belirgin değerlerin sayısıdır. Genel bir kural olarak, tüm olası farklı değerler sonlu ise, o zaman bu değişken kategorik bir değişken olarak değerlendirilebilir. Ancak, bu genel kural her zaman geçerli değildir. Yaş örneğini düşünün. Her ne kadar yaşın olası farklı değerleri sonlu olsa da, bazı durumlarda yaşı sürekli değişken olarak değerlendirmek daha mantıklı olabilir.

Kategorik değişkenlerin ayrıca iki alt tipi vardır:

### 2.1. Nominal değişkenler
Nominal değişkenler, kategorileri veya grupları temsil eder, böylece farklı değerler arasında sıralama olmaz. Örneğin, bir ürünün kaynak ülkesini gösteren bir değişkenimiz varsa, o zaman bu değişken yalnızca dünyadaki farklı ülke sayılarını alabilir. Dahası, bir nominal değişkenin alabileceği değerler arasında, sıra bir anlam ifade etmemektedir. Nominal bir değişkenin değerleri için büyüktür, küçüktür veya eşittir gibi ilişkiler açıkça tanımlanmamıştır. Ülkeyi temsil eden değişkenler, cinsiyet, tür vb. nominal değişkenlere örnek olarak verilebilir.

### 2.2. Ordinal değişkenler
Ordinal değişkenler sırayı gösterir. Örneğin, koşucular birincilik için 1, ikincilik için 2, üçüncülük için 3, vb. olarak skorlanabilir. Ancak, sıradaki bir değişken, skorlar arasındaki mesafeyle ilgili herhangi bir bilgi vermez.1′inci yer koşucusunun 2′nci yer koşucusundan daha hızlı olduğunu biliyoruz, ancak zamanlarındaki fark (değişken içindeki aralarındaki mesafe) dakika veya milisaniye olabilir. Ayrıca, 1′inci ile 2′nci arasındaki zaman aralığı arasındaki fark muhtemelen 2′nci ile 3′üncü arasındaki zaman farkı ile aynı değildir.

## 2. Aykırı değerlerin bulunması
Aykırı değerlerin tespiti, bunların ele alınmasında ilk adımdır. Bununla birlikte, bir aykırı değerin ne olduğunu tanımlamak için altın bir kural olmadığını vurgulamalıyız! Aykırı değerlerden bahsederken, onların nadir ve aşırı değerlere sahip olduklarından emin olmalıyız. Ancak, nadirlik ve aşırılığı tanımlamak öznel bir yargıdır ve üzerinde çalıştığımız işe bağlıdır. Aykırı değerlerin tanımlanması için yaygın olarak kullanılan bazı eşikler vardır.

Genellikle, aykırı değerleri tanımlamak için standart sapma kullanırız. (ortalama - Tstandart sapma, ortalama + Tstandart sapma) aralığının dışında kalan değerler aykırı değerlerdir. Burada T threasholdu ifade eder. Uygulamalara bağlı olarak, 1.5, 2, 3 veya 5'e eşit olarak tanımlanması yaygındır. Ancak, daha önce de söylediğimiz gibi, threashold için kesin değer elimizdeki işe bağlı olarak karar vermemiz gereken bir şeydir.

Bu kontrol noktasında, aykırı değerlerin tespit edilmesinin iki ana yolunu ele alıyoruz:

Değişkenleri görselleştirmek,
İstatistiksel teknikleri kullanmak.
İkinci yol daha bilimsel ve güvenilirdir ancak birincinin uygulanması kolaydır. Görselleştirme ile başlayalım.

### 2.1. Aykırı değerleri görsel olarak algılama
Verilerdeki aşırı değerleri tanımlamamıza yardımcı olan bazı faydalı grafikler vardır. Muhtemelen bu konuda en ünlü grafik kutu grafiğidir. Kutu-bıyık grafiği olarak da adlandırılan Kutu grafiklerinin çeyreklerarası aralıklara dayandığını ve bıyığı(whiskers) olduğunu unutmayın. Bıyığın ötesinde yatan noktalara fliers denir ve genellikle aykırı değer olarak kabul edilir. Matplotlib'de, varsayılan olarak, bıyıklar birinci ve üçüncü çeyreklerin uzağında 1.5*standart sapmaya varırlar, ancak bunu whis parametresini kullanarak kontrol edebiliriz. 

### 2.2. Aykırı değerlerin tespitinde istatistiksel yöntemler
### 2.2.1. Z-score
Bir değişkende aykırı değerler saptama yöntemlerinden birisi z-score dur. Z-score ilk önce verileri ölçeklendirir, böylece ortalaması 0 ve standart sapması 1 olur ve daha sonra bir değerin ortalamanın (0) ne kadar uzağında olduğunu ölçer. Bir değerin z-score'u ne kadar büyük olursa, o kadar büyük bir ihtimale aykırı değerdir. Bununla birlikte, bir değerin aykırı olduğunu söylemek için, z-score için bir eşik belirlememiz gerekir, bu sayede bu eşiğin üzerindeki skorların aykırı olduğu söylenir.

Scipy'ın zscore() işlevini kullanarak *goruntulenme_sayisi değişkeninin z-score'larını hesaplayalım ve eşik değerleri olarak 1, 2, 3 ve 4'ü ayarlayarak aykırı değerleri belirleyelim:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (8,5.5)
from scipy.stats import zscore
youtube = pd.read_csv("http://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/youtube.csv")
youtube.columns = ['sira', 'derece', 'kanal_adi', 'video_sayisi', 'abone_sayisi', 'goruntulenme_sayisi']
z_scores = zscore(youtube["goruntulenme_sayisi"])
for threshold in range(1,5):
    print("Eşik değeri: {}".format(threshold))
    print("Aykırı değerlerin sayısı: {}".format(len((np.where(z_scores > threshold)[0]))))
    print('------')
```

### 2.2.2. Tukey'in yöntemi
Bu yöntem Çeyrekler Arası Aralık (IQR/ÇAA) yöntemi olarak da bilinir ve kutu grafiği yöntemi hakkında konuştuğumuzda bu yöntemi çalışırken görürüz. Daha önce söylediğimiz gibi, ÇAA birinci ve üçüncü çeyrekler arasındaki aralıktır. Bu yöntem ÇAA'nın 1,5 katı dışında olan tüm değerleri aykırı değerler olarak değerlendirir. Ancak, aralığı değiştirebiliriz ve 1.5 yerine başka bir eşik değeri ayarlayabiliriz.

Eşik değerleri olarak 1, 1.5, 2, ...., 4.5 kullanarak, Tukey'in metoduyla goruntulenme_sayisi değişkenindeki aykırı değerlerin sayısını bulalım: Burada Numpy kütüphanesindeki .percentile() fonksiyonunu kullacağız. Bu fonksiyon bize yüzde olarak parametrelere denk gelen değerleri döndürür. Eşik değerlerini ve aykırı değerlerin sayısını bir veri çerçevesine atalım.

```python
q75, q25 = np.percentile(youtube["goruntulenme_sayisi"], [75 ,25])
caa = q75 - q25
```

```python
import pandas as pd
esik_degerleri = pd.DataFrame()

for esik_degeri in np.arange(1,5,0.5):
    min_deger = q25 - (caa*esik_degeri)
    maks_deger = q75 + (caa*esik_degeri)
    aykiri_deger_sayisi = len((np.where((youtube["goruntulenme_sayisi"] > maks_deger)| 
                                       (youtube["goruntulenme_sayisi"] < min_deger))[0]))
    esik_degerleri = esik_degerleri.append({'esik_degeri': esik_degeri,'aykiri_deger_sayısı': aykiri_deger_sayisi}, 
                                           ignore_index=True)
esik_degerleri
```

# 3. Aykırı değerler nasıl düzeltilir
Aykırı değerleri tespit ettikten sonra, analizimizin güvenilirliğini etkilememeleri için bunları ele almamız gerekir. Aykırı değerleri ele alırken genel olarak kullanılan üç farklı yaklaşım vardır:

Aykırı değerleri veri kümemizden çıkarabiliriz.

Aykırı değerlere yeni değerler atayarak sınırlayabiliriz.

Onları analizimiz için zararsız bir değere dönüştürürüz.

Şimdi, bu yaklaşımları tek tek gözden geçirelim.

## 3.1. Aykırı değerleri veri kümesinden çıkarma
Aşırı değerler taşıyan gözlemleri çıkarmak, aykırı değerleri ele almanın en kolay yoludur. Ancak, önceki dersteki veri kümesindeki herhangi bir gözlemi hariç tutarsak, bazı bilgileri etkili bir şekilde kaybedeceğimizi belirttiğimiz eksik değerler hakkındaki tartışmamızı hatırlayın. Bu, özellikle küçük veri kümelerindeki aykırı değerler için de geçerlidir. Bununla birlikte, aşağıdaki durumlar aykırı değerlerin veri kümesinden çıkarılmasını haklı kılabilir:

Aykırı değerlerin ölçüm hatasından veya veri toplama sürecindeki bir problemden kaynaklandığını biliyorsak, bunları çıkarmalıyız.

Eğer aykırı değerleri çıkarmak sonraki aşamalarda anlamlı bir değişiklik yapmıyorsa, onları çıkarabiliriz. Ancak, analizimizi hem aykırı değerlerlerle hem de aykırı değerler olmadan yapmamız gerekir.

Eğer aykırı değerleri çıkarmak takip eden aşamalarda anlamlı bir değişiklik yapıyorsa, ancak bunları ele almanın uygun bir yolunu bulamadıysak, onları veri kümemizden çıkarabiliriz.

## 3.2. Aykırı değerlerin sınırlandırılması
Aykırı değerlerin temel sorunu büyüklükleridir. Aykırı değerleri sınırlandırabilirsek, analizimiz üzerindeki etkileri sınırlı olacaktır. Aykırı değerleri sınırlandırmanın yaygın bir yoluna winsorization denir. winsorize etme tek yönlü veya iki yönlü olarak uygulanabilir. Tek yönlü winsorize etmede, ya değer dağılımının en düşük ucunu ya da en yüksek ucunu sınırlarız. İki yönlü de ise hem en düşük hem de en yüksek ucu sınırlandırırız.

winsorize etmede fikir, uç noktaların değerlerini belirli bir yüzdelik değere ayarlamaktır. Örneğin, dağıtımın en yüksek ucunun değerlerini 95 yüzdelik değerine ayarlayabiliriz. Benzer şekilde, en düşük ucun değerlerini 5 yüzdelik değerine ayarlayabiliriz.

Bir değişkeni scipy'ın winsorize() metodunu kullanarak winsorize edebiliriz. İki yönlü winsorize etmede parametre olarak ilk elemanın en düşük yüzdelik, ikinci elemanın ise en yüksek yüzdelik olduğu bir değer grubu verdik. Tek yönlü yüzdelik uygulamak istiyorsak, bu satırdaki ilgili öğeyi sıfıra(0) ayarlayabiliriz. Şimdi, goruntulenme_sayisi değişkenini winsorize edelim:

goruntulenme_sayisi değişkeninde aykırı değerlerimiz büyük değerlerde olduğu için tek yönlü olarak winsorize edebiliriz. Yüzde olarak da 90 yüzdelik dilimi kullanacağız.

```python
from scipy.stats.mstats import winsorize

winsorize_goruntulenme = winsorize(youtube["goruntulenme_sayisi"], (0, 0.10))

winsorize_goruntulenme
```

```python
plt.boxplot(winsorize_goruntulenme)
plt.title("Görüntülenme sayısının kutu grafiği (whis=1.5)")
plt.show()
```

Gördüğümüz gibi, kutu grafik artık aykırı değer içermiyor!

## 3.3. Aykırı değerleri dönüştürme
Aykırı değerlerle başa çıkmanın başka bir yaygın yolu da değişkenin dönüşümüdür. Dönüşüm, yalnızca aykırı değerlerde değil, aynı zamanda normal olmayan veya normale yakın olmayan dağıtımlarda da bize yardımcı olur. Değişkeni dönüştürürken, genellikle monotonik dönüşümler uygularız. Bu dönüşümler, bir değişkendeki değerlerin göreceli sırasını değiştirmezler. Log dönüşümü ve karekök dönüşümü veri biliminde en çok kullanılan dönüşümlerdir.

Log dönüşümünün goruntulenme_sayisi değişkeni üzerindeki etkisini görmek için, orijinal değişkenin ve log dönüşümü yapılmış versiyonunun histogramını çizelim:

Fakat şimdilik, log dönüşümünün aykırı değerler üzerinde bazı etkileri olup olmadığını görmek için orijinal değişkenin ve log dönüşüm değişkeninin kutu grafiklerine bakalım:

Çubuk grafiği kullanalım:

```python
plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
sns.barplot(youtube["derece"], youtube["winsorize_video_sayisi"])
plt.title("Ortalama Video Sayısı", fontdict = baslik_font)

plt.subplot(1,3,2)
sns.barplot(youtube["derece"], youtube["winsorize_aboneler"])
plt.title("Ortalama Abone Sayısı", fontdict = baslik_font)

plt.subplot(1,3,3)
# draw the heatmap using seaborn.
sns.barplot(youtube["derece"], youtube["winsorize_goruntulenme"])
plt.title("Ortalama Görüntülenme Sayısı", fontdict = baslik_font)

plt.show()
```

Gördüğümüz gibi, ortalamalar not azaldıkça azalır. Ancak, farklılıkları gözlemlemek, her bir kategori arasında bazı önemli farklılıklar olduğunu söylemenin resmi bir yolu değildir! Bunu yapmanın resmi yolu, t-testi veya ANOVA kullanmaktır!. Öyleyse, farklılıkların istatistiksel olarak anlamlı olup olmadığını öğrenmek için t-test kullanalım:

```python
# Test whether group differences are significant.
dereceler = youtube["derece"].unique()
grup_youtube = youtube.groupby("derece")
pd.options.display.float_format = '{:.15f}'.format
for var in ["winsorize_video_sayisi", "winsorize_aboneler", "winsorize_goruntulenme"]:
    karsilastirma = pd.DataFrame(columns=['grup_1', 'grup_2','istatistik', 'p_degeri'])
    print("{} için karşılaştırma".format(var),end='')
    for i in range(0, len(dereceler)):
        for j in range(i+1, len(dereceler)):
            ttest = stats.ttest_ind(youtube[youtube["derece"]==dereceler[i]][var], 
                                youtube[youtube["derece"]==dereceler[j]][var])
            grup_1 = dereceler[i]
            grup_2 = dereceler[j]
            istatistik = ttest[0]
            p_degeri = ttest[1]
            
            karsilastirma = karsilastirma.append({"grup_1" : grup_1 ,
                                                  "grup_2" : grup_2 ,
                                                  "istatistik": istatistik , 
                                                  "p_degeri" : p_degeri}, ignore_index=True)
    display(karsilastirma)
```

## 1.1. Kategorik değişkenlerle çalışmak
Çoğu model yalnızca sayısal veriler üzerinde çalışacağından, metin değerli kategorik verilerimizi sayısal bir değere dönüştürmemiz gerekir. Bu amaç için bir yöntem one-hot kodlama olarak adlandırılır. One-hot kodlamada, X seviyeli kategorik bir değişkenden bilgi edinmek için, her biri 1 veya 0 kodlanan X-1 adet sayısal özellik yaratırız. Her kategori için 0 veya 1 alan sonuç değişkenleri dummy veya gösterge değişkenler olarak bilinir. One-hot kodlamayı kategorik bir değişkene uygulamak için Pandas'ın .get_dummies() metodunu kullanabiliriz.

Youtube veri kümesini göz önünde bulundurun. derece değişkeni, 'A++', 'A+', 'A', 'A-', 'B+' değerlerini alan kategorik bir değişkendir. derece değişkeni beş düzeye sahipken, tüm bilgileri temsil etmek için yalnızca dört özellik gerekir: 'A++', 'A+', 'A' ve 'A-' değerlerinin 0 olduğunu biliyorsak, 'B+' değerinin 1 olması gerektiğini de biliriz. Özelliği olmayan seviyeye referans değeri adı verilir.

Youtube veri kümemizi yükleyelim ve derece değişkeni için dummy değişkenler oluşturalım:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from scipy.stats.mstats import winsorize
import warnings

warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
```

```python
youtube = pd.read_csv("http://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/youtube_winsorize.csv")
youtube.head()
```

```python
pd.get_dummies(youtube["derece"]).head()
```

.get_dummies() metodunun, gösterge değişkenleri derece değişkeninin değerleriyle adlandırarak her kategori için bir dummy değişken döndürdüğüne dikkat edin. Ancak, yukarıda açıkladığımız gibi onları bir modele dahil ettiğimizde dummy değişkenlerden birini çıkarmayı unutmamalıyız! Bu noktayı unutabileceğinizden korkuyorsanız, .get_dummies() metodunun drop_first parametresini True olarak ayarlayabilirsiniz. Bu, ilk kategoriyi çıkarır ve sadece geri kalanlar için gösterge değişkenler yaratır. Dahası, sadece pd.get_dummies(youtube["derece"]) i çalıştırmak, gösterge değişkenleri veri çerçevesinde tutmayacaktır. Bunu pd.concat() kullanarak yapabiliriz:

Pandas dersinden hatırlayacağınız gibi sutun bazında birleştirme yapacağımız için axis=1 parametresini ekliyoruz.

```python
youtube= pd.concat([youtube, pd.get_dummies(youtube["derece"])], axis=1)
youtube.head()
```

## 1.2. Normal dağılmayan bir değişken ile çalışmak
Daha önce de belirttiğimiz gibi, bazı makine öğrenme modelleri, hedef değişkenler için normal dağılımlar varsaymaktadır. Normallik bir model varsayımıysa, normal olmayan değişkenler bazen değişkenin kare kökü, ters veya log gibi dönüşümler kullanılarak normal dağılmış özelliklere dönüştürülebilir. Verilerin temizlenmesinin ve aykırı değerlerin tartışıldığı kontrol noktasında bu noktaya değindiğimizi hatırlayın. Log dönüşümü gibi dönüşümler uygulamak yalnızca aykırı değerlerle mücadelemize yardımcı olmakla kalmaz, aynı zamanda uygulandıkları değişkenlerin dağılımını da değiştirir.

Tek değişkenli analizden bahsettiğimiz kontrol noktasında, Youtube veri kümesindeki sürekli değişkenler için aşağıdaki grafiği göstermiştik:

```python
plt.figure(figsize=(18,15))


degiskenler = ['abone_sayisi', 'video_sayisi' , 'goruntulenme_sayisi']
winsorize_degiskenler = ['winsorize_aboneler', 'winsorize_video_sayisi', 'winsorize_goruntulenme']

for i in range(3):
    plt.subplot(3, 3, i+1)
    plt.hist(youtube[degiskenler[i]])
    plt.title(degiskenler[i], fontdict=baslik_font)
    
for i in range(3):
    plt.subplot(3, 3, i+4)
    plt.hist(youtube[winsorize_degiskenler[i]])
    plt.title(winsorize_degiskenler[i], fontdict=baslik_font)
    
for i in range(3):
    plt.subplot(3, 3, i+7)
    plt.hist(np.log(youtube[degiskenler[i]]))
    plt.title(degiskenler[i] + ' (log dönüşümlü)', fontdict=baslik_font)

plt.show()
```

Görünüşe göre, yukarıdaki üç değişkenin log dönüşümünün alınması, değişkenlerin dağılımını değiştirdi, böylece normal dağılım olmalarına rağmen, şimdi normal dağılıma daha fazla benziyorlar. Örneğimizde, log dönüştürülmüş değişkenlerin dağılımlarının scipy modülündeki Jarque-Bera ve Normal testlerini kullanarak hala normal dağılımı takip etmediğini istatistiksel olarak doğrulayabiliriz:

```python
from scipy.stats import jarque_bera
from scipy.stats import normaltest
pd.options.display.float_format = '{:.5f}'.format

ozellikler = ["abone_sayisi", "video_sayisi", "goruntulenme_sayisi"]
dagilim_testleri = pd.DataFrame(columns=['ozellik', 'jarque_bera_stats', 'jarque_bera_p_value', 
                                         'normal_stats', 'normal_p_value'])

for ozellik in ozellikler:
    jb_stats = jarque_bera(np.log(youtube[ozellik]))
    norm_stats = normaltest(np.log(youtube[ozellik]))
    dagilim_testleri = dagilim_testleri.append({"ozellik": ozellik,
                                                "jarque_bera_stats" : jb_stats[0] ,
                                                "jarque_bera_p_value" : jb_stats[1] ,
                                                "normal_stats": norm_stats[0] , 
                                                "normal_p_value" : norm_stats[1]
                                               }, ignore_index=True)
dagilim_testleri
```

P-değerleri sıfırdır ve bu da log dönüştürülmüş değişkenlerin hala normal dağılmadığını gösterir. Ancak dönüşümler yine de yararlı olabilir. Genelde normal dağılıma benzeyen dağılımlarla çalışmak (sol ve sağda iki kuyruk ve ortalamanın etrafında kümeleme değerleri), tek kuyruklu dağıtımlarla çalışmaktan daha kolaydır. Hatta bazı modellerde bu tür dağılımlar kullanılarak daha kolay elde edilebilecek yalnızca çok değişkenli normallik gereklidir. Dolayısıyla, bir değişkenin dağılımının bir dönüşümden sonra normal dağılıma daha yakın olduğunu görürsek, bu iyi olabilir.

# 1. Çok boyutluluk laneti (The curse of dimensionality)
Miyop olduğunu ve etrafındakileri görmek için gözlüğe ihtiyacınız olduğunu varsayalım. Bir gün uyanıyorsunuz ama dün gece uyumadan önce gözlüklerinizi nereye koyduğunuzu hatırlamıyorsunuz. Şimdi gözlüklerini arayalım:

Gözlüklerinizi sadece uzun ince bir masanın üzerine koyabileceğinizi hayal edin. Masanın genişliği sadece gözlüklerinizin genişliğindedir ve uzunluğu gözlüklerinizin uzunluğunun 100 katıdır. Gözlüklerinizi masada bulmak için en fazla 100 alana dokunmanız gerekir.

Bu kez masa yüzeyi bir kare olduğunu ve her kenarın uzunluğunun gözlüklerinizin uzunluğunun 100 katı olduğunu varsayalım. Şimdi, gözlüklerinizi bulmak için 
100∗100=10.000
 alana dokunmanız gerekir. Masanın boyutunu sadece 1'den 2'ye yükselttik ama en kötü durumda aramamız gereken alan sayısı 100 kat arttı!

Son olarak, masanın 3 boyutlu olduğunu düşünelim. Bir küp gibi hatta gözlükleri saklamak için raflarla dolu bir oda gibi de düşünebilirsiniz,! Her bir kenarın uzunluğu gözlüklerinizin uzunluğunun 100 katı ise 
100∗100∗100=1.000.000
 alana dokunmanız gerekir! Ne kadar çarpıcı bir artış!

Bu örnek size boyutluluk lanetinin ne olduğu konusunda bir fikir verir. Yarı-biçimsel olarak, çok boyutluluk laneti'ni, boyutların sayısındaki doğrusal bir artış için kombinasyon sayısındaki üssel artış olarak tanımlayabiliriz. Örnekte, masanın boyutunu 1'den 2'ye çıkarttığımızda, aramamız gereken alanların sayısının 100'den 10.000'e çıktığını gördük. Boyutu 2'den 3'e çıkardığımızda ise, toplam alan sayısı 10.000'den bir milyona çıktı.

Veri setimizdeki her değişken bir boyuttur. Yani, 20 özellikli bir model çalıştırırsak, girişimizin boyutunun 20 olduğunu ve 200 özelliğe sahip olduğumuzda girişin boyutunun 200 olduğunu söyleriz. İşleri biraz basitleştirelim ve Her boyut için optimum bir değer bulacak bir model çalıştırmayı düşünelim. Daha da basitleştirmek için, optimizasyon sürecinde her boyut için arayabileceğimiz sadece iki farklı değere sahip olduğumuzu varsayalım. Dolayısıyla sorunumuz, giriş boyutlarının her biri için iki değerden en iyisini bulmak. 20 özelliğimiz varsa, aramamız gereken toplam kombinasyon sayısı 
220=1048576'dır ve eğer 200 özelliğimiz varsa aramamız gereken sayı 
2200=16069380442589902755419620923411626025222029937827928330301376 olur!
İdeal olarak, veri setimizdeki her özelliği modellerimize dahil etmek isteyebiliriz. Fakat boyutluluk laneti yüzünden iki ciddi sorun bizi sınırlandırır:

Hesaplama süresi, modele her bir özellik eklendiğinde katlanarak artabilir. Bu nedenle, birçok özelliğe sahip modelleri çalıştırmak, çalıştırılması gereken süre nedeniyle olanaksız hale gelebilir!

Modellerimizi eğitmek için örnek veriler kullanıyoruz (bootcamp'ta gözetimli öğrenmeyi tartışmaya başladığımızda bunu daha iyi anlayacaksınız). Ancak, boyutluluk laneti nedeniyle, modellerimizi geliştirmek için gereken veri noktalarının sayısı, modele eklenen her özellik için katlanarak artmaktadır. Dolayısıyla, bu kadar veriyi bulmak imkansız hale gelebilir.

Bu iki ciddi nedenden ötürü, genellikle modellerimizdeki özelliklerin sayısını sınırlamaya çalışırız. Bunu yapabilmek için, veri kümesindeki tüm değişkenler arasından en kullanışlı özellikleri seçmemiz gerekir. Şimdi, özellik seçimi yaparken kullanılan genel yöntemleri tartışalım. Bu konuya geçmeden önce, boyutluluk laneti hakkında daha fazla kaynak okumak isterseniz, bu Wikipedia makalesi ile başlayın.

# 2. Özellik seçiminde yaklaşımlar
Özellik seçimi yaparken, sonuçla en güçlü bağlantıya sahip olan özellikleri korumak, aynı zamanda tabloya benzersiz bir şey getiren özelliklere de öncelik vermek isteriz. Hedefimiz seçenekleri yalnızca bir ideal özelliğe indirgemek değil, anlaşılması nispeten kolay olan, tahmin edilebilecek kadar güçlü, overfittingi en aza indiren ve nispeten etkili bir şekilde hesaplanabilir olan özellikler setine karar vermektir. Özellik seçimi, açıklayıcı güç ile model cimriliği arasında dengeleyici bir eylemdir. Neyse ki, veri bilimcilerin özellik kümelerini optimize etmelerine yardımcı olmak için birçok özellik seçim algoritması mevcuttur.

Özellik seçimi algoritmaları dört geniş gruba ayrılır. Bunlardan ikisi modelleme aşamasına entegre edilmiştir ve onları bootcamp'ın bu aşamasında değil, makine öğrenme algoritmalarını tartıştığımız sonraki modüllerde göstereceğiz. Ancak, daha önce bu bölümde tartıştığımız teknikleri kullanarak filtre yöntemlerini kolayca uygulayabilirsiniz. Bu derste bir boyut küçültme tekniği tanıtacağız.

## 2.1. Filtreleme yöntemleri (filtering methods)
Filtreleme yöntemleri, her bir özelliği ayrı ayrı değerlendirir, onlara özellikleri sıralamak için kullanılan bir "puan" verir. Belirli bir değerin üzerinde puanı olan özellikler tutulur veya atılır. Özellik, sonuçtan bağımsız olarak veya onunla birlikte değerlendirilebilir. Yalnızca belirli bir limitin üzerindeki varyansı olan özelliklerin tutulduğu varyans eşikleri, bağımsız olarak özelliklerin değerlendirilmesine bir örnektir. Her özelliğin sonuçla olan korelasyonu da filtre yöntemi olarak kullanılabilir ve bunu korelasyon tablosuna veya Pearson korelasyon matrisine bakarak yapabiliriz.

Filtre yöntemleri, hedefle ilgili olması muhtemel özellikleri seçmede iyidir. Bunlar kolay bir şekilde hesaplanırlar ve gayet açıktırlar, ancak özellikler arası ilişkiler göz önünde bulundurulmadığından fazladan özelliklerin listelerini üretme olasılıkları yüksektir. Çalıştırmaları "ucuz" olduklarından, sarıcı yöntemleri gibi daha fazla hesaplama gerektiren algoritmalar uygulamadan önce, filtreleme yöntemlerini azaltma özelliklerinde ilk geçiş olarak kullanabilirsiniz.

## 2.2. Sarıcı yöntemler (wrapper methods)
Sarıcı yöntemler, özellik kümelerini seçer. Bir modelde öngörülen güçleri açısından değerlendirilen ve performansı diğer kümelerin performansıyla karşılaştırılan farklı kümeler oluşturulur. Sarıcı yöntemler, özellik kümelerinin nasıl oluşturulduğuna göre farklılık gösterir. Bu iki özellik oluşturma yöntemi ileriye doğru geçiş ve geriye doğru geçiş şeklindedir. İleriye doğru geçişlerde, algoritma hiçbir özellik olmadan başlar ve özellikleri birer birer ekler, her zaman tahmin gücünde en yüksek artışa neden olan ve önceden belirlenmiş eşik değerinde duran özelliği ekler. Geriye doğru geçişlerde, algoritma tüm özelliklerle başlar ve özellikleri birer birer bırakır, her zaman en az tahmin gücü olan özelliği çıkarır ve önceden belirlenmiş bir eşikte durur. İleri ve geri geçiş yöntemleri açgözlü (greedy) olarak kabul edilir, çünkü bir özellik eklendiğinde (ileri) veya kaldırıldığında (geri) bir daha model için değerlendirilmez.

Sarıcı yöntemler, sonucu etkili bir şekilde öngören kullanışlı özellik kümelerinin seçilmesinde iyidir. Bununla birlikte, daha büyük özellik kümeleri için, bu yöntemler hesaplama açısından yoğun olabilir ve filtre yöntemlerinden daha fazla aşırı uyuma (overfitting) maruz kalabilir.

## 2.3. Gömülü yöntemler (embedded methods)
Gömülü yöntemler ayrıca özellik kümelerini de seçer. Ancak bunu, kullandığınız belirli model türü için uygun yöntemin özlü bir parçası olarak yapın. Bu, bir modelin öngörü gücünü değerlendirmek için kullanılan uygunluk ölçütlerine "karmaşıklık cezası"nın ilave edildiği düzenlileştirmeyi içerebilir. Gömülü yöntemler, sarıcı yöntemlerinin yararlarını sağlar, ancak hesaplama açısından daha az yoğundur. Farklı modeller farklı gömülü yöntemleri kullanır.

## 2.4. Boyut azaltma yöntemleri (dimensionality reduction methods)
Özellik seçiminde bir başka yaygın yaklaşım, boyut azaltma tekniklerini kullanmaktır. Bu yöntemler özellikle çok fazla özelliğe sahip olduğumuzda ve yukarıda bahsettiğimiz teknikleri uygulamak çok fazla hesaplama gerektiriyorsa faydalıdır. Çeşitli boyut küçültme teknikleri arasında, her veri bilimcinin bilmesi gereken popüler bir teknik vardır. Bu teknik Temel Bileşenler Analizi (Principal Components Analysis) ya da kısaca PCA olarak adlandırılır. Şimdi bu yöntemi inceleyelim.

# 3. Temel Bileşenler Analizi (Principal Components Analysis)
PCA, değişkenlerdeki bilgilerin çoğunu temsil eden daha küçük bir bileşen grubuna indirgemeye çalışan bir karmaşıklık azaltma tekniğidir. Kavramsal düzeyde PCA, varyansı paylaşan değişken kümelerini tanımlar ve bu varyansı temsil edecek bir bileşen yaratır. Örneğin, aşağıdaki iki görüntü, üç değişken arasında varyans paylaşmanın iki farklı yolunu gösterir. Her birinde, örtüşen alanlar paylaşılan varyansı temsil eder.



A kümesinin bir PCA'sı, muhtemelen, üç değişken tarafından da paylaşılan varyansı temsil eden ve dairelerdeki bilginin geri kalanını çıkaran bir bileşenle sonuçlanacaktır. Öte yandan B setinin bir PCA'sı, biri eflatun ve camgöbeği tarafından paylaşılan örtüşen alanı temsil eden ve diğeri eflatun/camgöbeği bileşeninde bulunmayan sarı renkteki varyansı temsil eden iki bileşene neden olacaktır. Her iki durumda da, biraz varyans kaybolmuştur.

Daha küçük bir özellik kümesi karşılığında varyansı kaybetmek uğraşmaya değebilir. Bazı model türleri (regresyon gibi), özelliklerin birbirleriyle ilişkisiz olacağını ve yüksek düzeylerde özellikler arası korelasyonun dengesiz çözümler ürettiğini varsaymaktadır. Daha az özellik içeren çözümler daha kolay anlaşılır ve hesaplama açısından daha verimlidir. Daha az özelliğe sahip olan çözümler overfittinge karşı da daha az hassastırlar.

## 3.1. İşler karışıyor
PCA normal dağılmış veriler için en iyi sonucu verir ve değişkenler arasındaki ilişkilerin doğrusal olduğunu varsayar. PCA ayrıca, değişkenlerin zayıf korelasyondan orta derecede güçlü korelasyona kadar değişmesi durumunda da en iyi sonucu verir. 0 korelasyon veya .7 veya .8'den büyük korelasyonlar problem yaratır. 0'a yakın korelasyonlar, paylaşılan hiçbir bilgi olmadığı anlamına gelir. Çok yüksek korelasyonlar, kararsız bileşenlere ve verilerdeki çok küçük değişikliklere cevabın çok fazla olduğu durumlara yol açabilir veya sadece bir veya iki değişkenli varyanstan oluşan bileşenlere neden olabilir. Birkaç çok yüksek veya çok düşük korelasyon iyidir. Çoğu zaman, diğer değişkenlerle çok yüksek veya çok düşük korelasyonlu bir değişkenimiz olduğunda problemler ortaya çıkabilir.

## 3.2. PCA: Uzayda dönme
n değişkenlerinin veri kümesini n boyutlu bir uzay olarak hayal edin. Bu, iki boyutlu bir uzayda iki değişkeni temsil eden, aşina olduğumuz iki boyutlu dağılım grafiğinin bir genellemesidir. PCA'da, değişkenler ilk önce ortalama 0 ve standart sapma 1 olacak şekilde standartlaştırılmıştır. Bu önemlidir çünkü şimdi tüm değişkenler başlangıç noktasından (tüm eksenlerin değerinin 0 olduğu) geçip aynı varyansı paylaşır. Daha sonra, n-boyutlu uzayın eksenleri, veri noktaları ve eksenler arasındaki mesafeyi minimize etmek için döndürülür.

Dönme sırasında, bazı eksenler kısalır ve bu, o eksen boyunca varyansın küçük olduğunu ve eksenin çok az bilgi içerdiğini gösterir. Bu eksen ve temsil ettiği ana bileşen, fazla bilgi kaybı olmadan çıkarılabilir.

## 3.3. PCA'nın Arkasındaki Matematik
Yukarıda gösterilen "dönme", verilerin kovaryans matrisini hesaplamaya ve doğrusal bir dönüşüm (dönme) uygulamasına ve bundan yeni eksenler tarafından açıklanan verilerdeki varyans miktarını ifade eden özdeğerlerin ve özvektörlerin türetilmesine dayanır. Bunu daha fazla araştırmak için biraz lineer cebire bakalım.

### 3.3.1. Kovaryans matrisi
Daha önce tartıştığımız bir korelasyon matrisi, kovaryansların varyanslara bölündüğü bir kovaryans matrisidir. Bu, kovaryansları standartlaştırır, böylece hepsi aynı ölçekte (-1 ila 1) olur ve karşılaştırılabilir. Kovaryans matrisleri, korelasyon matrisleri gibi, değişken çiftleri arasında paylaşılan varyans miktarı hakkında bilgi içerir.

x'in varyansı, x (xi) içindeki her bir değer ile x'in (¯x) ortalamasının kare farklarının toplamının örneklem büyüklüğüne (n) bölünmesiyle elde edilir.

var(x)=∑(xi−¯x)2n(Standart sapmanın √var(x) olduğuna dikat edin)
x ve y değişkenleri arasındaki arasındaki kovaryans, her değişken değeri ile ve bunun ortalamasının farklarının çarpımlarının örneklem büyüklüğüne bölünmesiyle elde edilir.

cov(A)=∑(xi−¯x)(yi−¯y)n

FIFA 19 verileri için kovaryans matrisini oluşturalım . Numpy kütüphanesindeki kovaryans matrisi verilerin satırlarla gösterilmesini gerektirir. Bu nedenle önce matrisimizin transpozunu alıyoruz:

### 3.3.2. Bir matris olarak veri
PCA aynı anda birçok değişkenle ilgilenir ve bu değişkenleri formüllerde m satır ve n sütunu içeren bir matris olarak gösterir; burada her satır bir ölçüm örneğini (örneğin, birçok değişken üzerindeki bir katılımcıdan gelen veriler) ve her sütun ise bir değişkeni temsil eder. Matematikte gelenek, matrislere büyük harflerle atıfta bulunulmasıdır. Örneğin, 3 değişkenli ve 4 katılımcı içeren bir veri kümesi 4 x 3 matris X olarak ifade edilir (satırların önce, sütunların sonra geldiğine dikkat edin - tıpkı Python'da bir veri çerçevesi, liste veya string dilimlenirken olduğu gibi).

### 3.3.3. PCA'ya dönelim
PCA'da amaç, kovaryans matrisini, tüm bilgilerin köşegenlerde (varyanslarda) olması ve hiçbirinin ters-köşegenlerde (kovaryanslarda) olmaması için dönüştürmektir. Bunu 
P matrisini bularak yaparız; burada P ve X'yi çarptığımızda, Y matrisini ve kapalı-köşegenlerinde sadece 0s olan CY kovaryans matrisini elde ederiz.

### 3.3.4. Özvektörler
Bir özvektör, bir bileşenin yönlü görünümüdür - önceki grafikteki kırmızı çizgidir. PCA sırasında, özvektörler ortogonal, yani 0 ile birbirleri arasında bir korelasyona sahip olacak şekilde seçilir. Aslında, bu sırayla yapılır. İlk önce, bu vektör ile veri noktaları arasındaki mesafeyi en aza indiren bir vektör bulunur. Bu vektör ilk bileşendir. Daha sonra, bu vektör ile veri noktaları arasındaki mesafeyi en aza indiren ikinci bir vektör bulunur, bu ikinci vektörün, uzayın n boyutlarından birinde ilk vektöre dik olması gerekir. Bu prosedür n adet vektör olana kadar devam eder.

### 3.3.5. Özdeğerler
Özdeğerler Özvektörlerin uzunluğunu temsil eder - her özvektörün bir özdeğeri vardır. Özvektörün uzunluğu, bir bileşen tarafından açıklanan toplam varyans oranını kodlar. Toplam varyans, PCA'daki değişkenlerin sayısına eşittir. Dolayısıyla, 1 değerinde bir Özdeğer, bileşenin bir değişkenle aynı miktarda varyansı açıkladığı anlamına gelir. 1'den büyük bir özdeğer, arzu edilen bir değerdir, çünkü 1 özdeğerli bir bileşen, herhangi bir bireysel değişkende yer alan bilgilerin ötesinde bir değer katmaz ve 1'den küçük bir özdeğer, kendi başına bir değişkenden daha az verimlidir. 2 değerinde bir özdeğer, bileşenin iki değişkene eşit miktarda bilgi içerdiği anlamına gelir. Tabii ki, bu bileşene yalnızca iki değişkenin yüklendiği anlamına gelmez.

## 3.4. Bileşenler
Bileşenler, özdeğerlerin karekökü ile bölünmüş özvektörlerdir. Bu, bileşenleri standartlaştırır, böylece bir bileşen tarafından açıklanan her değişkendeki varyans miktarı, bileşenler arasında karşılaştırılabilir. Veri azaltma amaçlarımız için, bileşenler faktörlü değişkenleri değiştirecektir çünkü belirli bir çözümün tüm bileşenleri birlikte değişkenlerin tüm paylaşılan varyansını temsil eder. PCA'nın yararı, değişkenlerden daha az bileşen bulunması ve böylece veri uzayını basitleştirmesidir. PCA'dan sonra, bileşenleri tam olarak değişkenleri kullandığımız gibi herhangi bir analizde kullanabiliriz.

Bileşenleri hesaplamak için FIFA 19'daki özvektörleri ve kovaryans matrisini kullanabiliriz, ancak önce bir karar vermeliyiz.

## 3.5. Kaç bileşen olmalı?
Bir PCA çalıştırılırken alınacak en büyük karar, kaç bileşenin saklanması gerektiğidir. PCA aslında korelasyon matrisindeki değişkenler kadar bileşeni bize geri verecektir. Eğer n değişkenimiz varsa ve n bileşeni tutmayı seçersek, orijinal verilerdeki bilgilerin %100'ünü yeniden üretebiliriz. Öte yandan, durumumuzu hiç basitleştirmeyeceğiz - hala değişkenler yerine bileşenler olarak ifade edilen aynı sayıda ayrı bilgi parçasıyla ilgileniyoruz.

Saklanacak bileşenlerin sayısını seçme konusunda bize yol gösterecek birkaç kural vardır. En açık olanı, özdeğerleri 1'den büyük olan bileşenleri değer katacağı için tutmaktır (çünkü tek bir değişkenden daha fazla bilgi içerirler). Bu kural ideal olandan daha fazla bileşen tutma eğilimindedir.

Diğer bir kural, özdeğerleri sırayla en yüksekten en düşüğe doğru bir çizgiyle bağlayarak görselleştirmektir. Buna “kayşat/scree” grafiği denir, çünkü sözde dağın dibinde biriken gevşek kayaya benzemektedir. Görsel incelemenin ardından analist, özdeğerleri çizginin eğiminin en şiddetli değiştiği noktaya ("dirsek" olarak da adlandırılır) düşen tüm bileşenleri tutacaktır.

Verideki varyansın sadece en az %x'ini açıklayan bileşenleri tuttuğumuz varyans kesmeleri ve PCA çözümünün eşdeğer rastgele veri üzerinde simüle edilmesini içeren programatik olanlar da dahil olmak üzere başka birçok kural da vardır. Nihayetinde, kaç tane bileşenin tutulacağı seçimi PCA yapma nedenlerinize bağlıdır.

Temel bir işlem bilgisi içermeyen bir metro navigasyon sistemi hayal edin: başlangıç noktanızı ve hedefinizi verirsiniz ve birbirleriyle nerede oldukları hakkında hiçbir bağlam olmadan kullanabileceğiniz durakların listesi verilir.

Londra Metro Haritası kadar verimli ve kullanışlı istatistiksel modeller oluşturmak her veri bilimcisinin amacıdır! İyi bir istatistik model yardımı ile istatistikçi olmayan birisi bile karmaşık bir veri setini anlayabilir. İstatistiksel bir model, bir veri kümesinin temel olmayan yönlerini gizleyerek, evrenin çalışmalarına küçük bir bakış atıyorsak güzel olabilir.
